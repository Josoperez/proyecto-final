# PROYECTO DE ANALISIS EXPLORATORIO DE DATOS CON PYTHON

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
```

## FASE UNO: CARGA Y VISUALIZACION DEL DATASET 

```python
data= pd.read_csv('Estadistico_Basica_y_Media_.csv')
display(data.head())
```

## FASE DOS: PROCESO DE LIMPIEZA DEL DATASET

#### ELIMINACIÓN DE DUPLICADOS

```python
print(f"\nNúmero de filas antes: {len(data)}")

data = data.drop_duplicates()

print(f"\nNúmero de filas después: {len(data)}")
```

#### REEMPLAZO DE VALORES FALTANTES

```python
valores_a_reemplazar = [ 'NO TIENE', 'NO APLICA', 'notiene@nose.com', 'no aplica', 
                       'SE ESCUENTRA  PRESTANDO SERVICOS EN LUGAR NO AUTORIZADO']

print(f"\nValores a reemplazar por NaN: {valores_a_reemplazar}")
print("\nConteo de valores faltantes antes:")
display(data.isna().sum())

data.replace(valores_a_reemplazar, np.nan, inplace=True)

print("\nConteo de valores faltantes después:")
display(data.isna().sum())
```

```python
data = data.drop(['FAX', 'EMAIL','WEB','BARRIO_VEREDA'], axis=1)
display(data.head())
```

```python
print("\nConteo de valores faltantes antes:")
display(data.isna().sum())
```

####  LIMPIEZA DE NOMBRES DE INSTITUCIONES

```python
display(data['NOMBRE_ESTABLECIMIENTO'].head())

data['NOMBRE_ESTABLECIMIENTO'] = data['NOMBRE_ESTABLECIMIENTO'].str.upper().str.strip()
data['NOMBRE_ESTABLECIMIENTO'] = data['NOMBRE_ESTABLECIMIENTO'].str.replace(r'\s+', ' ', regex=True)
```

#### CORRECCIÓN DE CÓDIGOS DANE

```python
print("\nCódigos antes (primeras 5):")
display(data[['COD_DANE_DEPARTAMENTO', 'COD_DANE_MUNICIPIO', 'CODIGO_DANE']].head())

data['COD_DANE_DEPARTAMENTO'] = data['COD_DANE_DEPARTAMENTO'].astype(str).str.zfill(2)
data['COD_DANE_MUNICIPIO'] = data['COD_DANE_MUNICIPIO'].astype(str).str.zfill(5)
data['CODIGO_DANE'] = data['CODIGO_DANE'].astype(str).str.strip()

print("\nCódigos después (primeras 5):")
display(data[['COD_DANE_DEPARTAMENTO', 'COD_DANE_MUNICIPIO', 'CODIGO_DANE']].head())
```

#### LIMPIEZA DE TELÉFONOS

```python
print("\nTeléfonos antes (primeras 5):")
display(data['TELEFONO'].head())

data['TELEFONO'] = data['TELEFONO'].astype(str).str.replace(r'[^0-9]', '', regex=True)
data['TELEFONO'] = data['TELEFONO'].replace('nan', np.nan)

print("\nTeléfonos después (primeras 5):")
display(data['TELEFONO'].head())
```

```python
display(data.head())
```
